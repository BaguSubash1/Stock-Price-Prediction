```python
data_label = data['Close', 'TSLA']
```

```python
data_label.head(3)
```

```python
#data_dimension = data.drop(columns=[('Close', 'TSLA')])
cols_to_select = [
    ('Open', 'TSLA'),
    ('day',  ''),
    ('month', ''),
    ('year', ''),
    ('dayofyear', ''),
    ('weekday', ''),
]

data_dimension = data[cols_to_select]


data_dimension.head(3)
```

```python
print(data_dimension.shape)
print(data_label.shape)
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(data_dimension,data_label,test_size = 0.2, shuffle = True)

```

```python
print(len(X_train))
print(len(Y_train))
print(len(X_test))
print(len(Y_test))
X_train.columns
      
```

```python
from sklearn.linear_model import LinearRegression 
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
y_predict = model.predict(X_test)

print(f"RMSE: {mean_squared_error(Y_test, y_predict):.2f}")
print(f"R² Score: {r2_score(Y_test, y_predict):.2f}")
```

```python
plt.figure(figsize=(10, 4))
plt.plot(Y_test.index, Y_test, label='Actual')
plt.plot(Y_test.index, y_predict, label='Predicted')
plt.title(f'{ticker} Stock Price Prediction')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
